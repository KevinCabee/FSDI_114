```python
# Simplified implementation of Queue (relying on built-ins)

class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, value):
        self.itmes.insert(0, value)

    def dequeue(self):
        return self.items.pop()

    # Nice to have methods
    def peek(self):
        return self.items[len(self.itmes)-1]

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []

```

```python
# From scratch implementation of QueueII

class QueueII:
    def __init__(self):
        self.front = None
        self.rear = None

    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def enqueue(self, value):
        new_node = self.__Node(value)
        if not self.rear:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
        

    def dequeue(self):
        if self.front:
            datum = self.front.data
            self.front = self.front.next
            return datum
        else:
            raise IndexError("The Queue is empty!")

# Nice to have methods

    def peek(self):
        if not self.front:
            raise IndexError("The Queue is empty!")
        else:
            return self.front.data


    def size(self):
        count = 0
        if self.front:
            current =  self.front
            while current:
                current  = current.next
                count += 1
            return count


    def is_empty(self): 
        return self.front == None
            
```

```python
def test(mytest):
    queue = QueueII()
    for char in mytest:
        queue.enqueue(char)
    print("The Queue contains %s elements" % queue.size())
    out = ""
    while not queue.is_empty():
        out += queue.dequeue()
    return out

print(test("Kevin"))
```
